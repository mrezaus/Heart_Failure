```python
model = tf.keras.models.Sequential([
    tf.keras.Input(shape=(144,)),
    tf.keras.layers.Dense(64),
    tf.keras.layers.Dense(16),
    tf.keras.layers.Dense(1,'sigmoid')
])
```

```python
model = tf.keras.models.Sequential([
    tf.keras.Input(shape=(144,)),
    tf.keras.layers.Dense(16),
    tf.keras.layers.Dense(8),
    tf.keras.layers.Dense(1,'sigmoid')
])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(lr=3e-7), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
```
